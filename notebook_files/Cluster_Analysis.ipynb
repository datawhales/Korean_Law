```python
label_dict = dict()

for i in range(len(df)):
    label = df.loc[i]['조항구분']
    if str(label) != 'nan':
        label = label[1:-1]
        label = re.sub("'", '', label)
        label_list = label.split(',')
        label_list = [label.strip() for label in label_list] 
        for label in label_list:
            if label not in label_dict:
                label_dict[label] = 1
            else:
                label_dict[label] += 1
```

```python
from collections import Counter

def use_keyword(df, *keyword):
    counter = Counter()
    for i in range(len(df)):
        try:
            for k in keyword:
                if k in str(df.iloc[i]['원본']):
                    counter.update({str(df.iloc[i]['조항구분']): 1})
                    break
            
        except:
            pass
    return counter.most_common(), sum(counter.values())
```

```python
use_keyword(df, '날짜')
```

```python
def keyword_count_for_each_label(df, *keyword):
    each_label_cnt_dict = {k:v for k, v in zip(key_list, [0 for _ in range(len(key_list))])}
    cnt_list, _ = use_keyword(df, *keyword)
    for label, cnt in cnt_list:
        labels = [x.strip() for x in re.sub("'", "", label[1:-1]).split(',')]
        for l in labels:
            if l in each_label_cnt_dict:
                each_label_cnt_dict[l] += cnt
    whole_dict = {k:v for k, v in zip(key_list, [0 for _ in range(len(key_list))])}
    for i, (v1, v2) in enumerate(zip(each_label_cnt_dict.values(), ordered_label_dict.values())):
        whole_dict[key_list[i]] = (v1, v2)
    return whole_dict
```

```python
def target_label_keyword_count(df, target_label, *keyword):
    each_label_cnt_dict = {k:v for k, v in zip(key_list, [0 for _ in range(len(key_list))])}
    cnt_list, _ = use_keyword(df, *keyword)
    num_of_wrong_labels = 0
    for label, cnt in cnt_list:
        labels = [x.strip() for x in re.sub("'", "", label[1:-1]).split(',')]
        if target_label in labels:
            each_label_cnt_dict[target_label] += cnt
        elif labels != ['a']:
            num_of_wrong_labels += cnt
            for l in labels:
                if l in each_label_cnt_dict:
                    each_label_cnt_dict[l] += cnt
    whole_dict = {k:v for k, v in zip(key_list, [0 for _ in range(len(key_list))])}
    for i, (v1, v2) in enumerate(zip(each_label_cnt_dict.values(), ordered_label_dict.values())):
        whole_dict[key_list[i]] = (v1, v2)
    print(f"Percentage of correct: {whole_dict[target_label][0] / whole_dict[target_label][1] * 100:.4f}")
    print(f"Number of data that have keywords and belong to target label: {whole_dict[target_label][0]}")
    print(f"Number of data that have keywords but not belong to target label: {num_of_wrong_labels}")
    return whole_dict
```

```python
def keyword_count_for_each_label_by_line(df, *keyword):
    each_label_cnt_dict = {k:v for k, v in zip(key_list, [0 for _ in range(len(key_list))])}
    cnt_list, _ = use_keyword(df, *keyword)
    for label, cnt in cnt_list:
        if label in each_label_cnt_dict:
            each_label_cnt_dict[label] += cnt
    
    whole_dict = {k:v for k, v in zip(key_list, [0 for _ in range(len(key_list))])}
    for i, (v1, v2) in enumerate(zip(each_label_cnt_dict.values(), ordered_label_by_line_dict.values())):
        whole_dict[key_list[i]] = (v1, v2), int(v1 / v2 * 100)
    return whole_dict
```
