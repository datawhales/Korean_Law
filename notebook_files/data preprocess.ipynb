## ['필수조항8', 'nan'] -> ['필수조항8'] 처리

```python
import numpy as np
import ast 

def transform_label(label_string):
    new_label = []
    if str(label_string) != 'nan':
        label_list = ast.literal_eval(label_string)
        for l in label_list:
            if l != 'nan':
                new_label.append(l)
    return str(new_label) if new_label else np.nan
```

```python
df['조항구분'] = df['조항구분'].apply(transform_label)
```
